### Python Wrappers

```python
base_url='http://api.weatherapi.com/v1'
api_method='/current.json'
api_key='6cd41fedd40349868cd194344230410'
city='Tallahassee'

request_url=f"{base_url}{api_method}?key={api_key}&q={city}"
print(request_url)
```

```python
res=requests.get(request_url)
weather_data=res.json()
weather_data
```

```python
current_temp=weather_data['current']['temp_f']
feels_like=weather_data['current']['feelslike_f']
current_temp
feels_like
```

```python
class CityWeather:
    def __init__(self, name, region, current, feels_like, condition):
        self.city=name
        self.region=region
        self.current_temp=current
        self.feels_like=feels_like
        self.condition=condition
        
    def __str__(self):
        return f"It is currently {self.condition}"

class WeatherAPI:
    base_url='http://api.weatherapi.com/v1'
    def __init__(self, api_key):
        self.api_key=api_key
    #underscore is a naming convention to signal this is a private method    
    def _get(self, city, api_method, **params):
        request_url=f"{self.base_url}{api_method}?key={self.api_key}&q{city}"
        response=requests.get(request_url)
        if response.ok:
            data=response.json()
            return data
        else:
            return None
    
    def get_current_weather(self, city):
        self._get(city, '/current.json')
```
